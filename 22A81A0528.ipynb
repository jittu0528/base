```python
import numpy as np
arr=np.array([1,2,3,4,5])
print(arr)
a=np.zeros((3,3),dtype=int)
print(a)
b=np.ones((2,2),dtype=int)
print(b)
a_arr=np.arange(10)
print(a_arr)
reshape=arr.reshape(5,1)
print(reshape)
sli=arr[2:4]  #slicing
print(sli)
a1=np.array([1,2,3,4,5])
b1=np.array([11,12,13,14,15])
c=np.vstack(a1+b1)  #vertical stack
d=np.stack(a1+b1)   #horizontal
print(c)
print(d)
ar=np.array([1,2,3,4,5,6])
split=np.split(ar,3)    #split
print(split)
tr=np.array([[1,2,3,4],[4,5,6,7]]) #transpose
trans=tr.T
print(trans)
```

```python
#linear algebra
import numpy as np
a=np.array([[1,2],[4,3]])
b=np.array([[2,4],[5,2]])
c=np.dot(a,b)
print(c)
d=np.linalg.eig(c)
print(d)
```

```python
a=[230,560,780,127,128]
b=[200,160,270,127,400]
years=[1,2,3,4]
profit_a=[(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b=[(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='hotpink',linewidth='3',label='companyA',marker='>',ms='10',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label='companyB',marker='H')
plt.show()
```
